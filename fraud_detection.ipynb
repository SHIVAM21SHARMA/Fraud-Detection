DATA PREPROCESSING

```
#preprocessing
from sklearn import preprocessing
for f in X.columns:
    if X[f].dtype=='object' :
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(X[f].values))
        X[f] = lbl.transform(list(X[f].values))     
```

FEATURE SCALING

```
from sklearn.preprocessing import StandardScaler
SC_X=StandardScaler().fit_transform(X)
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=0.99999)
PC_X = pca.fit_transform(SC_X)
```

PERFORMING SAMPLING TECHNIQUES TO SOLVE THE DATA IMBALANCY PROBLEM

```
#solving class imbalancy problem
#using undersampling method
from imblearn.under_sampling import NearMiss
```

```
# implementing undersampling for handling imbalanced
nM = NearMiss(random_state=42)
X_res,y_res=nM.fit_resample(PC_X,y)
```

FEATURE SELECTION USING AUTOMATIC BACKWARD ELIMINATION

```
#try
# FEATURE SELECTION
#Automatic implementations of Backward Elimination in Python with p-value
import statsmodels.api as sm
def backwardElimination(y,x, sl):
    numVars = len(x[0])
    numRows = len(x)
    x = np.append(arr = np.ones( (numRows,1) ).astype(int), values=x, axis=1)
    for i in range(0, numVars):
        print(i,end=' ')
        numVars = len(x[0])
        regressor_OLS = sm.OLS(y, x).fit()
        pVal=regressor_OLS.pvalues.astype(float)
        maxVar = max(pVal)
        if maxVar > sl:
            for j in range(0, numVars):
                if (pVal[j] == maxVar):
                    x = np.delete(x, j, 1)
                    break;
        else:
            break
    print('sup')
    regressor_OLS.summary()
    return x
x_opt=backwardElimination(y_res,X_res,0.05)
```

SPLITTING THE DATASET INTO TRAINING AND TEST SET

```
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_opt, y_res, test_size = 0.2, random_state = 0)

```

LOGISTIC REGRESSION MODEL

FITTING THE TRAINING SET INTO LOGISTIC REGRESSION MODEL

```
# ALGORITHM 1
# classification  -> logistic regression

from sklearn import  linear_model,metrics 

     
# create logistic regression object 
logReg = linear_model.LogisticRegression(max_iter=100000)

   
# train the model using the training sets 
logReg.fit(X_train,y_train)
  
# making predictions on the testing set
y_pred =logReg.predict(X_test) 
   

```

CONFUSION MATRIX WITH ACCURACY AND LOSS LOG

```
# comparing actual response values (y_test) with predicted response values (y_pred) 

from sklearn.metrics import confusion_matrix

print("confusion metrics =")
confuse= confusion_matrix(y_test,y_pred)
print(confuse)

print("Logistic Regression model accuracy(in %):", metrics.accuracy_score(y_test,y_pred)*100)

print("loss =\n",metrics.log_loss(y_test,y_pred))

```

```
from sklearn.metrics import f1_score
print(f1_score(y_test,y_pred))
```

SVM MODEL

```
# ALGORITHM 2
#svm classification
# Fitting SVM to the Training set
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

```
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
print("confusion metrics =")
confuse= confusion_matrix(y_test,y_pred)
print(confuse)

print("SVM model accuracy(in %):", metrics.accuracy_score(y_test,y_pred)*100)

print("loss =\n",metrics.log_loss(y_test,y_pred))
```
